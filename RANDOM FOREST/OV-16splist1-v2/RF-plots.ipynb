```python
cdf = crossdf[crossdf["features"]==0.33]
cdf.head(2)
```

```python
bestallruns  = [] 
for species in range(16):
    sp0 = cdf[cdf["sp"]==species]
    for ridx in range(5):
        rundx = sp0[sp0["runidx"]==ridx]
        bestidx = rundx["nnseagg"].idxmax()
        bestallruns.append([species,ridx,rundx.T[bestidx]["max_depth"],rundx.T[bestidx]["trees"],rundx.T[bestidx]["features"],rundx.T[bestidx]["nnseavg"],rundx.T[bestidx]["nnseagg"]])
```

```python
bestdf =  pd.DataFrame(bestallruns,columns=["species","runidx","depth","trees","feats","nnseavg","nnseagg"])
```

```python
bestdfv2 = []
for species in range(16):
    newdf = bestdf[bestdf["species"]==species]
    modevalmd = stats.mode(newdf["depth"])[0][0]
    modevalfreqmd = stats.mode(newdf["depth"])[1][0] 
    modevaltrees =  stats.mode(newdf["trees"])[0][0]
    modevalfreqtrees =  stats.mode(newdf["trees"])[1][0]
    modevalfeats = stats.mode(newdf["feats"])[0][0]
    modevalfreqfeats = stats.mode(newdf["feats"])[1][0]
    nnseaggmean = newdf["nnseagg"].mean()
    nnseaggstd = newdf["nnseagg"].std()
    bestdfv2.append([species,modevalmd,modevalfreqmd,modevaltrees,modevalfreqtrees,modevalfeats,modevalfreqfeats,nnseaggmean,nnseaggstd])
    
```

```python
bestRF = pd.DataFrame(bestdfv2,columns=["species","modeMD","freqMD","modetr","freqtr","modefeats","freqfeats","nnseaggMEAN","nnseaggSTD"])
```
