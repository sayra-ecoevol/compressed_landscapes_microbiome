```python
import pandas as pd
import numpy as np
import cvxpy as cvx
import random
import time


B = pd.read_pickle("~/bge-analysis-simv3/B.pkl")
A = pd.read_pickle("~/bge-analysis-simv3/A.pkl")


import os


from sklearn.metrics import r2_score



from sklearn.metrics import mean_absolute_percentage_error as mape
from scipy import stats
```

```python
tableg=A.T
species_ones = [] 
for species in range(16):
    species_ones.append(tableg[tableg[species]==1.0][species].index.to_list())
```

```python
samplist = [327]
```

```python
max_depthl = [2,5,10,15,20]
ntrees = [100,200,300,400,500,1000]
max_featuresl = ["auto","sqrt","log2"]
```

```python
data = pd.read_pickle("~/compressed_sensingv1/realdatasets/BENCHMARKING/RANDOM FOREST/synthetic dataset/repAllRFRELotherthan1p.pkl")
```

```python
rep9 = data[data[1]==9]
samp = rep9[rep9[0]==32]
samp[samp[3]==10]
```

```python
len(max_depthl)*len(ntrees)*len(max_featuresl)
```

```python
df = [] 
for replica in range(10):
    rdfread = pd.read_pickle("~/compressed_sensingv1/realdatasets/BENCHMARKING/RANDOM FOREST/synthetic dataset/rep"+str(replica)+"RFREL.pkl")
    df.append(rdfread)
```

```python
rdff = pd.concat([df[0],df[1],df[2],df[3],df[4],df[5],df[6],df[7],df[8],df[9]])
```

```python
per = [] 
for species in range(16):
    spdff = rdff[rdff[1]==species]
    for replica in range(10):
        repdf = spdff[spdff[0]==replica]
        steadystate = pd.read_pickle("~/compressed_sensingv1/old-data/fullsearch/15_2.3_"+str(replica)+"/steady-state.pkl")
        n = 2**16
        sst = steadystate.T
        sstv2 = sst.copy()
        sstv2["total"] = sstv2.T.sum()
        for i in range(16):
            sstv2[i] = sstv2[i]/sstv2["total"]
        sstv3 = sstv2.fillna(0)
        abunvec = pd.DataFrame(sstv3[species].values[species_ones[species]])
        abunvec.index = np.arange(0,2**15,1)
        for m in samplist:
            ri = pd.read_pickle("~/compressed_sensingv1/abun_simv3/training-data/sp"+str(species)+"_rep"+str(replica)+"-in-sample-"+str(m)+".pkl")    
            onescomms =  tableg.T[species_ones[species]].T.index 
            actcomms = pd.DataFrame(onescomms).T[ri].T[0].values
            
            tedata = list(set(list(np.arange(0,2**15,1))).difference(ri))
            tecomms = pd.DataFrame(onescomms).T[tedata].T[0].values
            yte = abunvec.T[tedata].T[0].values
            for md in max_depthl:
                mdf = repdf[repdf[2]==md]
                for n_est in ntrees:
                    treesdf = mdf[mdf[3]==n_est]
                    for mfeatures in max_featuresl:
                        mfeats = treesdf[treesdf[4]==mfeatures]
                        recon = pd.DataFrame(mfeats[5].iloc[0])
                        ybestdf1 = pd.DataFrame(recon)
                        ybestdf1[ybestdf1[0]<1e-6]=0
                        slope, intercept,r_value, p_value, std_err = stats.linregress(yte,ybestdf1[0].values)
                        rval2 = r2_score(yte,ybestdf1[0].values)
                        gop2 = 1/(2-rval2)
                        mape2 = mape(yte,ybestdf1[0].values)
                        per.append([replica,species,md,n_est,mfeatures,slope,intercept,r_value**2,rval2,gop2,mape2])
```

```python
pdf = pd.DataFrame(per,columns=["replica","species","max_depth","trees","features","slope","intercept","r2bestfit","r2","nnse","mape"])
pdf.to_pickle("RF-per-REL.pkl")
```

```python
mfeatsc = pdf[pdf["features"]=="sqrt"]
mdvar = mfeatsc[mfeatsc["trees"]==100]
```

```python
md_choose_best = [] 
for replica in range(10):
    repdf = mdvar[mdvar["replica"]==replica]
    for species in range(16):
        spdf = repdf[repdf["species"]==species]
        md_choose_best.append([replica,species,spdf["nnse"].max()])
```

```python
mdchdf = pd.DataFrame(md_choose_best)
```

```python
mdchdf.to_pickle("RF_for_maxdepth_per.pkl")
```
